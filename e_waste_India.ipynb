## Data cleaning

### Subtask:
Clean the loaded dataset `df` by handling missing values, outliers, and potential duplicates, ensuring data consistency and correctness.


**Reasoning**:
I need to start by checking for missing values in the DataFrame `df` and handle them based on an appropriate strategy. I will then check for duplicate rows and remove them if any are found. Finally, I'll print the data type information of each column. These steps are directly related to the instructions: "Missing Values Handling," "Duplicate Rows Handling," and "Data Consistency Check".



```python
import numpy as np

# Check for missing values
print("Missing values per column:\n", df.isnull().sum())

# Strategy for handling missing values:
# The column 'Any other thoughts on E-waste' has many missing values, which are likely because people have nothing else to add.
# We can fill these missing values with a string like 'No further comments'.
df['Any other thoughts on E-waste'].fillna('No further comments', inplace=True)
#check if there is still null value
print("\nMissing values per column after filling:\n", df.isnull().sum())

# Check for duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Confirm that duplicates are removed
print("\nNumber of duplicate rows after removal:", df.duplicated().sum())

# Examine data types
print("\nData types:\n", df.dtypes)
```

**Reasoning**:
Based on the previous output, there are still two missing values in the column "How likely are you to participate in an e-waste recycling program if it was convenient and free?". As there are only two missing data and this is a categorical column, I will drop these two rows. Additionally, there are no numerical columns to perform outlier detection, so I will skip that part, and I will finish by printing the data information using the `.info()` method, as required in instruction 5.



```python
# Drop rows with missing values in 'How likely are you to participate in an e-waste recycling program if it was convenient and free?'
df.dropna(subset=['How likely are you to participate in an e-waste recycling program if it was convenient and free?'], inplace=True)

# Print the data information
df.info()
```

## Data wrangling

### Subtask:
Transform the raw data in the DataFrame `df` to make it more suitable for in-depth analysis, focusing on preparing the data for potential modeling or more detailed analysis.


**Reasoning**:
I should start by converting the Likert scale responses to numerical values as instructed in the first and second steps. I will create a mapping dictionary and apply it to the specified columns. This will involve iterating through the target columns and using the `.replace()` method to perform the conversion.



```python
likert_mapping = {
    'Strongly Disagree': 1,
    'Disagree': 2,
    'Neutral': 3,
    'Agree': 4,
    'Strongly Agree': 5,
    'Not at all': 1,
    'Slightly': 2,
    'Moderately': 3,
    'Very': 4,
    'Extremely': 5,
    'Never': 1,
    'Rarely': 2,
    'Sometimes': 3,
    'Often': 4,
    'Very often': 5,
    'Not very likely': 1,
    'Somewhat likely': 2,
    'Likely': 3,
    'Very likely': 4,
    'Extremely likely': 5,
    'Not important': 1,
    'Slightly important': 2,
    'Neutral': 3,
    'Important': 4,
    'Very important': 5,
    'Very dissatisfied': 1,
    'Dissatisfied': 2,
    'Neutral': 3,
    'Satisfied': 4,
    'Very satisfied': 5,
    'Very difficult': 1,
    'Difficult': 2,
    'Neutral': 3,
    'Easy': 4,
    'Very easy': 5,
    'Less than once a year': 1,
    'Once a year': 2,
    'A few times a year': 3,
    'Once every few months': 4,
    'Once a month': 5,
    'Somewhat aware': 2,
    'Unaware': 1,
    'Aware': 3,
    'Very aware':4,
    'Extremely aware':5,
}

columns_to_convert = [
    "How aware are you of the environmental impact of improper e-waste disposal?",
    "How likely are you to participate in an e-waste recycling program if it was convenient and free?",
    "How important is it for you that your e-waste is disposed of responsibly?",
    "How satisfied are you with the e-waste management services in your area?",
    "How often do you dispose of electronic devices (e.g., phones, computers, appliances)?",
]

for col in columns_to_convert:
    df[col] = df[col].replace(likert_mapping)
```

**Reasoning**:
Now that the initial Likert scale columns have been converted, I need to verify the conversion by checking the data types and value counts as stated in step 3. After that, I will handle the remaining categorical columns, simplifying them if needed as instructed in step 4, and then re-check the data information in step 5.



```python
# Verify the conversion and check data types
print("Data types after conversion:")
print(df[columns_to_convert].dtypes)

# Check value counts for each converted column
print("\nValue counts after conversion:")
for col in columns_to_convert:
    print(f"\nValue counts for column: {col}")
    display(df[col].value_counts())

# Handle other categorical columns
print("\nHandling other categorical columns:")
df['Age'] = df['Age'].replace({'Under 18': '18-24', '55 or older': '55+'})
df['Occupation'] = df['Occupation'].replace({'Retired': 'Other', 'Student': 'Other', 'Unemployed': 'Other'})
df['Gender'] = df['Gender'].replace({'Other': 'Prefer not to say'})
df['How do you segregate your waste?'] = df['How do you segregate your waste?'].replace({'I do not segregate': 'No', 'Dry, wet': 'Yes'})

# Check the data again
print("\nData information after handling other categorical columns:")
df.info()
```

**Reasoning**:
From the previous output, we can see that some columns are not converted to numeric values correctly, so we need to fix that. Specifically, "How often do you dispose of electronic devices (e.g., phones, computers, appliances)?" and "How likely are you to participate in an e-waste recycling program if it was convenient and free?" are still objects. We will need to add the values to the mapping dictionary and apply it again. After that, the task is completed, we can check the data information one more time, and then report the success status.



```python
likert_mapping.update({
    'More than twice a year': 6,
    'Twice a year': 5,
    'Unlikely':0
})

columns_to_convert = [
    "How often do you dispose of electronic devices (e.g., phones, computers, appliances)?",
    "How likely are you to participate in an e-waste recycling program if it was convenient and free?"
]

for col in columns_to_convert:
    df[col] = df[col].replace(likert_mapping)

# Check the data information again
print("\nData information after the second conversion:")
df.info()
```

Conduct in-depth data analysis on the transformed DataFrame `df` to identify key trends, patterns, and insights related to e-waste management in India.

The subtask is to conduct in-depth data analysis on the transformed DataFrame `df`. This involves calculating descriptive statistics, analyzing distributions, investigating relationships between variables, identifying correlations, and summarizing findings. I will start by calculating descriptive statistics for the numerical columns and analyzing the distribution of responses in the categorical columns, as instructed in steps 1 and 2.



```python
import pandas as pd

# Calculate descriptive statistics for numerical columns
numerical_cols = [
    "How often do you dispose of electronic devices (e.g., phones, computers, appliances)?",
    "How aware are you of the environmental impact of improper e-waste disposal?",
    "How satisfied are you with the e-waste management services in your area?",
    "How likely are you to participate in an e-waste recycling program if it was convenient and free?",
    "How important is it for you that your e-waste is disposed of responsibly?"
]
print("Descriptive Statistics for Numerical Columns:")
display(df[numerical_cols].describe())

# Analyze the distribution of responses in categorical columns
categorical_cols = [col for col in df.columns if col not in numerical_cols]
print("\nDistribution of Responses in Categorical Columns:")
for col in categorical_cols:
    print(f"\nValue counts for column: {col}")
    display(df[col].value_counts())
```

**Reasoning**:
I have completed steps 1 and 2 of the instructions. Now, I will proceed with step 3, which involves investigating the relationship between different variables using `.groupby()`. I will group the data by `Age` and calculate the mean of the numerical columns within each group, as per the example provided in the instructions. Then, I will perform step 4, calculating the correlations between the numerical variables and printing the result.



```python
# Investigate the relationship between different variables using groupby
print("\nRelationship between Age and Numerical Variables:")
display(df.groupby('Age')[numerical_cols].mean())

# Identify correlations between numerical variables
print("\nCorrelations between Numerical Variables:")
correlation_matrix = df[numerical_cols].corr()
display(correlation_matrix)
```

```python
# Summarize the findings
print("\nKey Insights:")
print("- E-waste Disposal Frequency: The mean disposal frequency of electronic devices is around 2.5 times per year, with variations across age groups. The 45-54 age group disposes of electronic devices slightly more frequently than other age groups.")
print("- Environmental Awareness: On average, respondents have a moderate level of awareness regarding the environmental impact of improper e-waste disposal, with a mean score of around 2.58 out of 4. The awareness levels are quite similar across different age groups.")
print("- Satisfaction with E-waste Management: The average satisfaction level with e-waste management services is moderate, with a mean score of around 2.84 out of 5. Different age groups show similar levels of satisfaction.")
print("- Willingness to Participate in Recycling Programs: The average likelihood of participating in e-waste recycling programs if they were convenient and free is moderately high, with a mean score of 3.20 out of 4. The 35-44 age group shows the highest likelihood, while the 45-54 age group shows the lowest.")
print("- Importance of Responsible Disposal: Respondents generally consider it very important that their e-waste is disposed of responsibly, with a high mean score of approximately 4.39 out of 5. The 45-54 age group shows the highest concern.")
print("- Categorical Variable Analysis: Most respondents (80 out of 133) know that e-waste refers to discarded electronic devices. About half of the respondents use informal recyclers or municipal waste collection for e-waste disposal. Environmental concerns are the primary motivation for most respondents to recycle e-waste. Lack of information or local facilities are the main barriers for proper e-waste disposal.")
print("- Correlation Analysis: There is a moderate positive correlation (0.38) between the likelihood of participating in recycling programs and the importance of responsible disposal. There is a low positive correlation (0.27) between awareness of environmental impact and satisfaction with e-waste management. There is a low negative correlation (-0.22) between how often the respondent dispose of electronic devices and how satisfied the respondent is about the e-waste management.")
print("- Age Group Insights: The 45-54 age group shows the highest concern for responsible disposal and disposes of devices slightly more frequently. The 35-44 age group shows the highest willingness to participate in recycling programs.")
```

## Data visualization

### Subtask:
Create informative visualizations to represent the key findings of the analysis on e-waste management in India, using the data in the DataFrame `df`.


**Reasoning**:
The subtask is to create visualizations based on the analysis performed on the `df` dataframe. Since there are multiple bar charts and one heatmap to create, I will create all the charts in one code block for better code structure and efficiency. The instructions are very clear, and I will implement every step with code.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar chart for "What do you think e-waste means?"
plt.figure(figsize=(12, 6))
sns.countplot(y='Do you know anything about E-waste', data=df, palette='viridis')
plt.title('Distribution of Responses: What do you think e-waste means?')
plt.xlabel('Count')
plt.ylabel('Response')
plt.tight_layout()
plt.show()

# 2. Bar chart for "How do you dispose of your e-waste?"
plt.figure(figsize=(12, 6))
sns.countplot(y='What method do you primarily use to dispose of your e-waste?', data=df, palette='viridis')
plt.title('Distribution of E-waste Disposal Methods')
plt.xlabel('Count')
plt.ylabel('Disposal Method')
plt.tight_layout()
plt.show()

# 3. Bar chart for average disposal frequency by age group
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='How often do you dispose of electronic devices (e.g., phones, computers, appliances)?', data=df, palette='viridis')
plt.title('Average Disposal Frequency of Electronic Devices by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Disposal Frequency')
plt.tight_layout()
plt.show()

# 4. Bar chart for average awareness of environmental impact by age group
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='How aware are you of the environmental impact of improper e-waste disposal?', data=df, palette='viridis')
plt.title('Average Awareness of Environmental Impact by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Awareness Score')
plt.tight_layout()
plt.show()

# 5. Bar chart for average satisfaction level by age group
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='How satisfied are you with the e-waste management services in your area?', data=df, palette='viridis')
plt.title('Average Satisfaction Level with E-waste Management Services by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Satisfaction Score')
plt.tight_layout()
plt.show()

# 6. Bar chart for average likelihood of participating in recycling programs by age group
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='How likely are you to participate in an e-waste recycling program if it was convenient and free?', data=df, palette='viridis')
plt.title('Average Likelihood of Participating in E-waste Recycling Programs by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Likelihood Score')
plt.tight_layout()
plt.show()

# 7. Bar chart for average importance of responsible disposal by age group
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='How important is it for you that your e-waste is disposed of responsibly?', data=df, palette='viridis')
plt.title('Average Importance of Responsible Disposal by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Importance Score')
plt.tight_layout()
plt.show()

# 8. Heatmap for correlation matrix of numerical variables
numerical_cols = [
    "How often do you dispose of electronic devices (e.g., phones, computers, appliances)?",
    "How aware are you of the environmental impact of improper e-waste disposal?",
    "How satisfied are you with the e-waste management services in your area?",
    "How likely are you to participate in an e-waste recycling program if it was convenient and free?",
    "How important is it for you that your e-waste is disposed of responsibly?"
]
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.tight_layout()
plt.show()
```
